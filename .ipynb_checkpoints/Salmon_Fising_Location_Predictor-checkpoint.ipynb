```python
def Stats(data, year):
    for yr in year:
        m = data[yr].mean()
        s = data[yr].std()
        v = data[yr].var()
        sk = data[yr].skew()
        
        print(yr)
        print('Mean =', m)
        print('Standard Deviation =', s)
        print('Variance =', v)
        print('Skew =', sk)
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov

def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr

def SpearmanCorr(xs, ys):
    xs = pd.Series(xs)
    ys = pd.Series(ys)
    return xs.corr(ys, method='spearman')
```

```python
def RelationBetween(x, y):
    print('Covariance =', Cov(x, y))
    print("Pearson's Correlation =", Corr(x, y))
    print('Spearman Correlation =', SpearmanCorr(x, y))
```

```python
RelationBetween(data['Area_num'], data['2017'])
```

```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
corr_data_1 = data.Area_num, data.Species_num
ht_area_spec = CorrelationPermute(corr_data_1)
pvalue_1 = ht_area_spec.PValue()
pvalue_1
```

Our P-value returned is 0 meaning we are not able to reject the null hypothesis that there is no correlation between the Area and the Species reported.
