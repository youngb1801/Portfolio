A scenario where one might be interested in determining if texit is positive or negative is when parsing reviews. I have choosen a dataset from Kaggle.com on Women’s Clothing E-Commerce reviews. We will look at the reviews and determine if the review is positive, neutral or negative.

```python
# Analyzing text for whether comments are positive or negative

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np

Addr1='/Users/bretyoung/Desktop/DSC 550/Womens Clothing E-Commerce Reviews.csv'
dataframe = pd.read_csv(Addr1)
print(dataframe)
corpus = dataframe['Review Text'].dropna()

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print("")
print("vectorized words")
print("")
print(vectorizer.get_feature_names())
print("")
print("Identify Feature Words - Matrix View")
print("")
print( X.toarray())

df = pd.DataFrame({'text' : corpus})
```

```python
# a list of positive and negative words were found here: 
# https://github.com/shekhargulati/sentiment-analysis-python/blob/master/opinion-lexicon-English/positive-words.txt

file_pos = open('/Users/bretyoung/Desktop/DSC 550/positive-words.txt', 'r')


# create a list of positive words
positive_list = file_pos.read().split('\n')
positive_list = positive_list[35:]
print(positive_list)
```

```python
file_neg = open('/Users/bretyoung/Desktop/DSC 550/negative-words.txt', 'r',encoding="ISO-8859-1")

# create a list of negative words
negative_list = file_neg.read().split('\n')
negative_list = negative_list[35:]
print(negative_list)
```

```python
# loop through words to see if they are in positive or negative lists

df['positive'] = 0 
df['negative'] = 0
df['TotScore'] = 0

for i in range(0, len(df)):
    try:
        for j in df.text.str.split()[i]:
            if j in positive_list:
                df.positive[i] += 1 # add 1 for each positive word
            elif j in negative_list:
                df.negative[i] += 1 # add 1 for each negative word
    except KeyError:
         continue
    df.TotScore[i] = df.positive[i] - df.negative[i] # total positive minus negative

print(df)

Z = sum(df.TotScore)

print("")
print("Overall Score: ",Z)

```
